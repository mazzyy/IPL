```python
from collections import Counter
# if a team wins till count number

# count to check
check_count = 3

# initialize a counter for the names
name_counts = Counter()

# loop over the names and update the counter
for name in inverted_df['winner']:
    name_counts[name] += 1
    if name_counts[name] == check_count:
        break
#         

# print the final counts for each name
for name, count in name_counts.items():
    print(f'{name}: {count}')
```

```python
#  set the maximum number of iterations
max_iterations = 59

# initialize a dictionary to store the counts
name_counts = {}

# loop over the names and update the counts
for i, name in enumerate(inverted_df['winner'].loc[61:]):
    if i == max_iterations:
        break
    if name not in name_counts:
        name_counts[name] = 0
    name_counts[name] += 1

# print the final counts for each name
for name, count in name_counts.items():
    print(f'{name}: {count}')

# find the name with the highest count
winner = max(name_counts, key=name_counts.get)
print(f'The winner is {winner} with {name_counts[winner]} counts.')
```

```python
score_year_winner = pd.DataFrame(columns=['heigest_matches_winner', 'year', 'winning_count','tournament_winner'])

for year in inverted_df['year'].unique():
    win_count = inverted_df[inverted_df['year'] == year].groupby('winner').size()
    name = win_count.idxmax()
    winning_count = win_count.max()
    score_year_winner = score_year_winner.append({
        'heigest_matches_winner': name,
        'year': year,
        'winning_count': winning_count
    }, ignore_index=True)
```

```python
winners = {
    2008: 'Rajasthan Royals',
    2009: 'Deccan Chargers',
    2010: 'Chennai Super Kings',
    2011: 'Chennai Super Kings',
    2012: 'Kolkata Knight Riders',
    2013: 'Mumbai Indians',
    2014: 'Kolkata Knight Riders',
    2015: 'Mumbai Indians',
    2016: 'Sunrisers Hyderabad',
    2017: 'Mumbai Indians',
    2018: 'Chennai Super Kings',
    2019: 'Mumbai Indians',
    2020: 'Mumbai Indians',
    2021: 'Chennai Super Kings',
    2022: 'Gujarat Titans'
}
for year in score_year_winner['year']:
    score_year_winner.loc[score_year_winner['year'] == year, 'tournament_winner'] = winners[year]
    
    
```

```python
for y in inverted_df['year'].unique():
    n_rows = len(inverted_df[inverted_df['year'] == y])
    winner = winners[y]
    inverted_df.loc[inverted_df['year'] == y, 'tournament_winner'] = [winner]*n_rows
```
