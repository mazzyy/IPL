```python
df['team1_score']=df['team1_score'].isna().mean()
df['team2_score']=df['team2_score'].isna().mean()
```

```python

df.drop(['stadium','full_scorecard','place','man_of_the_match'],axis=1,inplace=True)
```

```python
df=df.dropna()
```

```python
df.shape
```

```python
df
```

```python
# use str.split() to split the name column into first_name and last_name columns
df[['first_name', 'last_name']] = df['margin'].str.split(' ', expand=True)
# df_num = pd.get_dummies(df) 
```

```python
# create a boolean mask to check if the score contains "wickets"
mask = df['score'].str.contains('wickets')

# use the boolean mask to assign values to two new columns
df.loc[mask, 'wickets'] = df.loc[mask, 'score']
df.loc[~mask, 'runs'] = df.loc[~mask, 'score']
```

```python
df['wickets']=df['wickets'].str.split(' ', expand=True)[0]
df['runs']=df['runs'].str.split(' ', expand=True)[0]
```
